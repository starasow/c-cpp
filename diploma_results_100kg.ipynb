```python
import numpy as np
import matplotlib.pyplot as plt 

fig, ax = plt.subplots()  
YES = [[0.002, 0.3], [0.002, 1.5], [0.002, 1.0],[0.002, 0.6], [0.002, 0.45],[0.002, 0.35], [0.002, 0.2], [0.01, 1.5], [0.01, 1.0], [0.01, 0.6]]
NO = [[0.02, 0.2], [0.01, 0.2], [0.005, 0.2], [0.0025, 0.2], [0.02, 0.3], [0.02, 0.5], [0.02, 0.8], [0.01, 0.4], [0.02, 1.5]]
for i in range(len(YES)):
    ax.scatter(YES[i][0], YES[i][1], color='green')
for i in range(len(NO)):
    ax.scatter(NO[i][0], NO[i][1], color='red')

ax.set_title('Начальные условия для малого спутника M = 100kg', fontsize=14)
plt.xlabel(r'$\dfrac{v}{a}$', fontsize=16, rotation=0)
plt.ylabel(r'$\dfrac{\rho S L}{m}$     ', fontsize=16, rotation=0)
ax.tick_params(axis='both', labelsize=12)
plt.subplots_adjust(top=1.0, bottom=0.00, left=0.0, right=1.0, hspace=0.1, wspace=0.5)
ax.grid()
fig.savefig("100kg.pdf")
plt.show()
```
